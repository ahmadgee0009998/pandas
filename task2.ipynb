```python
import numpy as np
import pandas as pd
df_customer=pd.read_csv("customers.csv")
df_sales=pd.read_csv("sales.csv")
df_sales
```

```python
df_customer.info()
df_sales.info()
df_sales.duplicated().sum()
df_sales["date"]=pd.to_datetime(df_sales["date"])
df_sales.info()
```

```python
df_sales.groupby("region")["sales"].sum()
df_sales.groupby("customer_type")["sales"].sum()
```

```python
df_merged=pd.merge(df_sales,df_customer,on="order_id")
df_merged.columns
```

```python
df_merged=pd.pivot_table(df_merged,values="sales",index="region",columns="customer_type",aggfunc="mean")
df_merged
```
